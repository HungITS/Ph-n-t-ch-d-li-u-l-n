```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_crawl.csv')
df
```

```python
df = df.dropna(how='all')
```

```python
df
```

```python
def normalize_price(price):
    if pd.isna(price):  # Nếu giá trị là NaN
        return None
    if isinstance(price, str):  # Chỉ xử lý nếu giá trị là chuỗi
        price = price.replace(',', '.').strip()  # Đổi dấu phẩy thành dấu chấm và bỏ khoảng trắng
        if 'tỷ' in price:
            return float(price.replace(' tỷ', ''))  # Loại bỏ từ "tỷ" và chuyển sang số thực
        elif 'triệu' in price:
            return float(price.replace(' triệu', '')) / 1000  # Chuyển từ triệu sang tỷ
    return None  # Trả về None nếu không hợp lệ

def normalize_price_m2(price_m2):
    if pd.isna(price_m2):  # Nếu giá trị là NaN
        return None
    if isinstance(price_m2, str):  # Chỉ xử lý nếu giá trị là chuỗi
        price_m2 = price_m2.replace(',', '.').strip()  # Đổi dấu phẩy thành dấu chấm và bỏ khoảng trắng
        if 'triệu' in price_m2:
            return float(price_m2.replace(' triệu/m²', ''))  # Loại bỏ từ "tỷ" và chuyển sang số thực
        elif 'đ/m²' in price_m2:
            return float(price_m2.replace(' đ/m²', '').replace('.', '')) / 1000  # Chuyển nghin sang trieu
    return None  # Trả về None nếu không hợp lệ
df1 = df.copy()
df1['Gia'] = df1['Gia'].apply(normalize_price)
df1['Gia/m2'] = df1['Gia/m2'].apply(normalize_price_m2)
df1['DienTich'] = round(df1['Gia'] * 1000 / df1['Gia/m2'], 0)
```

```python
df1
```

```python
# Có thể thấy 2 trường dữ liệu Gia/m2 và Gia tương đương nhau nên mình sẽ xóa trường này đi
df1 = df1.drop(columns=["Gia/m2"])
```

```python
df1
```

```python
# Lay so phong ngu (bo chu 'phong')
def clean_phongngu(value):
    if 'nhiều hơn' in value:
        return float(value.split(' ')[-2]) + 1
    else:
        try:
            return float(value.split(' ')[0])
        except ValueError:
            return None  # Giá trị không hợp lệ
    
# Xử lý dữ liệu trong cột "Phongngu"
df2 = df1.copy()
df2["Phongngu"] = df2["Phongngu"].apply(clean_phongngu)
```

```python
# Lay so phong tam (bo chu 'phong')
def clean_phongtam(value):
    # Kiểm tra nếu giá trị là NaN hoặc không phải chuỗi
    if pd.isna(value) or not isinstance(value, str) or value.strip() == '' or 'hơn' in value.lower():
        return None  # Hoặc giá trị mặc định
    if 'Nhiều hơn' in value:
        return float(value.split(' ')[2]) + 1
    try:
        return float(value.split(' ')[0])
    except ValueError:
        return None
    
df3 = df2.copy()
df3["PhongTam"] = df3["PhongTam"].apply(clean_phongtam)
```

```python
# Lay du lieu Quan
def extract_quan(dia_chi):
    if isinstance(dia_chi, str):  # Kiểm tra nếu "DiaChi" là chuỗi
        parts = dia_chi.split(',')  # Tách chuỗi bằng dấu phẩy
        if len(parts) >= 2:  # Kiểm tra nếu có ít nhất 2 phần tử
            return parts[-2].strip()  # Lấy phần tử thứ hai từ cuối và loại bỏ khoảng trắng
    return None  # Trả về None nếu không phải chuỗi hoặc không hợp lệ
df4 = df3.copy()
df4['Quan'] = df4['DiaChi'].apply(extract_quan)

df4 = df4[df4['Quan'].str.contains('Quận|Huyện', na=False)]

df4 = df4.reset_index(drop=True)
```

```python
frame = df4[['DiaChi','Quan','TinhTrangBDS','DienTich','Phongngu','TenPhanKhu','SoTang','PhongTam','Loai','GiayTo','MaCanHo'
             ,'TinhTrangNoiThat' ,'HuongCuaChinh','HuongBanCong','DacDiem',"Gia"]]
# Chinh lai kieu du lieu so cho cac cot so
frame['DienTich'] = frame['DienTich'].replace('', np.nan).astype('float')
frame['Phongngu'] = frame['Phongngu'].replace('', np.nan).astype('float')
frame['PhongTam'] = frame['PhongTam'].replace('', np.nan).astype('float')
frame['Gia'] = frame['Gia'].replace('', np.nan).astype('float')
```

```python
# Chinh lai kieu du lieu so cho cac cot so
frame['DienTich'] = frame['DienTich'].replace('', np.nan).astype('float')
frame['Phongngu'] = frame['Phongngu'].replace('', np.nan).astype('float')
frame['PhongTam'] = frame['PhongTam'].replace('', np.nan).astype('float')
frame['Gia'] = frame['Gia'].replace('', np.nan).astype('float')
```

```python
frame.to_csv('dataset.csv',encoding="utf-8-sig",index=False)
```
